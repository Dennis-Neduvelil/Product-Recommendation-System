```python
import pandas as pd
import random
```

```python
df=pd.read_csv('ratings.csv')
df
```

```python
transactions_list=[]
for i in range(100000):
    transactions_list.append(i+200000)
transactions_list
    
```

```python
product_id=list(df['id'])
product_id
```

```python
ratings_list=[]
for i in range(100000):
    ratings_list.append(random.randint(0,5))
ratings_list
```

```python
products_list=[]
for i in range(100000):
    products_list.append(random.choice(product_id))
products_list
```

```python
users=[]
for i in range(1000):
    users.append(i+1000)
users
```

```python
users_list=[]
for i in range(100000):
    users_list.append(random.choice(users))
users_list
```

```python
transactions=pd.Series(transactions_list,name='Transactions_id')
users=pd.Series(users_list,name='User_id')
products=pd.Series(products_list,name='Products')
ratings=pd.Series(ratings_list,name='Ratings')
```

```python
new_dff=pd.concat([transactions,users,products,ratings],axis=1)
new_dff
```

```python
new_dff.info()
```

```python
new_dff.to_csv('transactions.csv', header=True, index=False)
print('File Saved')
```
