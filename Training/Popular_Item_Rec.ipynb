```python
ratings_list=[]
for i in range(44446):
    ratings_list.append(random.randint(1,5))
    
```

```python
ratings_list
```

```python
reviews_number=[]
for i in range(44446):
    reviews_number.append(random.randint(1,5000))
reviews_number
```

```python
ratings=pd.Series(ratings_list,name='Ratings')
ratings
```

```python
reviews=pd.Series(reviews_number,name='Reviews')
reviews
```

```python
id=df['id']
id
```

```python
df=pd.concat([id,ratings,reviews],axis=1)
df['Rank']=df['Ratings']*df['Reviews']
df
```

```python
popular_df=df[df['Ratings']>=4].sort_values('Rank',ascending=False)
popular_df.head(50)
```

```python
popular_df.to_csv('popular.csv', header=True, index=False)
print('File Saved')
```
